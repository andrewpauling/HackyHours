### Example of how to regrid from hybrid-sigma to pressure coordinates using new xgcm functionality

Install xgcm with `conda install -c conda-forge xgcm`

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import xarray as xr
from xgcm import Grid
import os

mpl.rcParams['figure.figsize'] = 12, 8
mpl.rcParams['font.size'] = 18
```

### Load some example model output to use

```python
ddir = '/glade/work/apauling/CCSM1degcontoutput'
dfile = 'CCSM41degcont.cam.h0.T.0071-0100avg.nc'
ncf = os.path.join(ddir, dfile)

ds = xr.open_dataset(ncf)
ds
```

### Define function to regrid to specified pressure levels

This calculates the pressure at every grid point in `(time, lev, lat, lon)` space, then uses the new  
`xgcm.Grid.transform` method to interpolate the data onto constant pressure levels based on the pressure coordinate we just created

```python
def regrid_xgcm(ds, var):
    # Calculate the pressure at every time, lev, lat, lon
    p = (ds['hyam']*ds['P0'] + ds['hybm']*ds['PS'])/100
    # Assign it as a variable in the dataset
    ds = ds.assign({'p': np.log(p)})
    
    # Create an xgcm Grid object with a Z coordinate given by 'lev'
    grid = Grid(ds, coords={'Z': {'center': 'lev'}}, periodic=False)
    
    # Give the array of pressures to interpolate to
    p_target = np.array([10., 20., 30., 50., 70., 100., 150., 200., 250.,
                         300., 400., 500., 600., 650., 700., 750., 800., 850.,
                         900., 950., 1000.])
    
    # Use the transform method to interpolate to constant pressure given the target pressures above
    # The target_data parameter tells it what variable to use to base the transformation on. 
    # In our case, we're using the model pressure at every point calculated above
    varout = grid.transform(ds[var], 'Z', np.log(p_target), target_data=ds.p)
    
    # Rename the new dimension and assign the coordinate as the target pressures above
    varout = varout.rename({'p': 'plev'})
    varout = varout.assign_coords({'plev': p_target})
    
    return varout.squeeze()
```

```python
# Get the data on hybrid-sigma levels and pressure levels
t_hybridsigma = ds['T'].squeeze()
t_pressure = regrid_xgcm(ds, 'T')
```

### Plot the results

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
t_hybridsigma.mean('lon').squeeze().plot.pcolormesh('lat', 'lev', ax=ax1)
t_pressure.mean('lon').squeeze().plot.pcolormesh('lat', 'plev', ax=ax2)

ax1.set_title('Hybrid-sigma coordinate')
ax2.set_title('Pressure coordinate')

ax1.invert_yaxis()
ax2.invert_yaxis()
```
